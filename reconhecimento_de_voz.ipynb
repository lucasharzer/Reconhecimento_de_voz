```python
import speech_recognition as sr
from time import sleep


def linha():
    print('\033[1;35m----------------------------------------\033[m')

def titulo(msg):
    print(msg.center(50))
    
linha()
titulo('\033[1;34mReconhecimento de Fala\033[m')
linha()
sleep(2)

# r é reconhecimento
r = sr.Recognizer()

print('\n\033[1;34mEssas são as opções de microfone no seu computador: \033[m')
linha()
sleep(2)
print(sr.Microphone().list_microphone_names())
sleep(2)

while True:
    linha()
    resposta = str(input('\033[1;34mVocê está pronto para falar[S/N]? \033[m')).strip().upper()
    if resposta == "S":
        # m é microfone

        linha()
        print("\n\033[1;34mAguarde ...\033[m\n")
        sleep(2)

        with sr.Microphone() as m:
            r.adjust_for_ambient_noise(m)
            linha()
            print('\033[1;34mPode falar que eu vou gravar: \033[m')
            linha()
            audio = r.listen(m)
            texto = r.recognize_google(audio, language="pt-BR")
            print(f'\n-> \033[1;32m{texto}\033[m')
        
        linha()
        terminar = str(input('\033[1;34mVocê deseja Sair[S/N]? \033[m')).strip().upper()
        if terminar == 'S':
            break

print('\n\033[1;34mAguarde ...\033[m\n')
sleep(2)
print("\033[1;32mFinalizado!\033[m")
```
